```python
# Simple Calculator using numbers for operators

count = int(input("How many calculations do you want to perform? "))

for i in range(count):
    print("\nCalculation", i + 1)

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    print("Choose an operator:")
    print("1 = Plus (+)")
    print("2 = Minus (-)")
    print("3 = Multiply (*)")
    print("4 = Divide (/)")

    choice = int(input("Enter operator number (1-4): "))

    if choice == 1:
        print("Result:", num1 + num2)

    elif choice == 2:
        print("Result:", num1 - num2)

    elif choice == 3:
        print("Result:", num1 * num2)

    elif choice == 4:
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
        else:
            print("Result:", num1 / num2)

    else:
        print("Error: Invalid operator number.")
3
```
